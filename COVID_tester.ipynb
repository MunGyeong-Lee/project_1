```python
import tkinter as tk

root = tk.Tk()
root.title('코로나 진단 키특 예측도 계산 프로그램')
root.geometry('600x500+100+100')

def make_label14():
    lab14.config(text=ent12.get())

def make_label24():
    lab24.config(text=ent22.get())    

def make_label34():
    lab34.config(text=ent32.get()) 

def make_ppv_npv():
    sen=float(lab14.cget("text"))/100
    spe=float(lab24.cget("text"))/100
    P_R=float(lab34.cget("text"))/100

    PPV=round((P_R*sen)/(P_R*sen+(1-P_R)*(1-spe))*100, 2) 
    NPV=round(((1-P_R)*spe)/((1-P_R)*spe+P_R*(1-sen))*100,2)

    lab52.configure(text=PPV)
    lab62.configure(text=NPV)    


#1행 라벨
lab11=tk.Label(root, text="민감도",
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab11.grid(row=0,column=0,padx=5,pady=10)

ent12=tk.Entry(bg='white', width=8, font=('맑은 고딕',16,'bold'))
ent12.grid(row=0,column=1,padx=5,pady=10)

button13=tk.Button(root, text='\u2192',font=('맑은고딕',12,'bold'),bg='red',fg='white',width=4, height=1, command=make_label14)
button13.grid(row=0,column=2,padx=5,pady=10)

lab14=tk.Label(root, 
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab14.grid(row=0,column=3,padx=5,pady=10)


#2행 라벨
lab21=tk.Label(root, text="특이도",
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab21.grid(row=1,column=0,padx=5,pady=10)

ent22=tk.Entry(bg='white', width=8, font=('맑은 고딕',16,'bold'))
ent22.grid(row=1,column=1,padx=5,pady=10)

button23=tk.Button(root, text='\u2192',font=('맑은고딕',12,'bold'),bg='red',fg='white',width=4, height=1, command=make_label24)
button23.grid(row=1,column=2,padx=5,pady=10)

lab24=tk.Label(root, 
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab24.grid(row=1,column=3,padx=5,pady=10)


#3행 라벨
lab31=tk.Label(root, text="발생률",
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab31.grid(row=2,column=0,padx=5,pady=10)

ent32=tk.Entry(bg='white', width=8, font=('맑은 고딕',16,'bold'))
ent32.grid(row=2,column=1,padx=5,pady=10)

button33=tk.Button(root, text='\u2192',font=('맑은고딕',12,'bold'),bg='red',fg='white',width=4, height=1, command=make_label34)
button33.grid(row=2,column=2,padx=5,pady=10)

lab34=tk.Label(root, 
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab34.grid(row=2,column=3,padx=5,pady=10)


#4행 라벨
button41=tk.Button(root, text='결과계산하기(클릭)',font=('맑은고딕',16,'bold'),bg='red',fg='white',width=8, height=1, command=make_ppv_npv)
button41.grid(row=3,column=0,padx=5,pady=10, columnspan=4, sticky='we')


#5행 라벨
lab51=tk.Label(root, text="양성예측도(양성인데 진짜 걸림)",
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab51.grid(row=4,column=0,padx=5,pady=10, columnspan=3, sticky='we')

lab52=tk.Label(root, 
                bg='yellow',fg='black',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab52.grid(row=4,column=3,padx=5,pady=10)


#6행 라벨
lab61=tk.Label(root, text="음성예측도(음성인데 진짜 안걸림)",
                bg='#2F5597',fg='white',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'))
lab61.grid(row=5,column=0,padx=5,pady=10, columnspan=3, sticky='we')

lab62=tk.Label(root, 
                bg='yellow',fg='black',
                width=8, height=1,
                font=('맑은 고딕',16,'bold'),
                )
lab62.grid(row=5,column=3,padx=5,pady=10)



root.mainloop()
```
